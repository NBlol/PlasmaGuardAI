```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")

def visualize_full_dataset(data):
    # Plot 1: Plasma Current Comparison (Reference vs Reconstruction)
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='reference_time', y='reference_plasma_current', data=data, label='Reference Plasma Current', color='blue')
    sns.lineplot(x='reference_time', y='reconstruction_plasma_current', data=data, label='Reconstruction Plasma Current', color='red')
    plt.title('Reference vs Reconstruction Plasma Current over Time')
    plt.xlabel('Time')
    plt.ylabel('Plasma Current')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot 2: Z-Axis Reference vs Reconstruction
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='reference_axis_z', y='reconstruction_axis_z', data=data, color='green')
    plt.title('Reconstruction Axis Z vs Reference Axis Z')
    plt.xlabel('Reference Axis Z')
    plt.ylabel('Reconstruction Axis Z')
    plt.grid(True)
    plt.show()

    # Plot 3: X-Point Z Reference vs X-Point Z Reconstruction
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='x_point_z_reference', y='x_point_z_reconstruction', data=data, color='purple')
    plt.title('X Point Z Reference vs X Point Z Reconstruction')
    plt.xlabel('X Point Z Reference')
    plt.ylabel('X Point Z Reconstruction')
    plt.grid(True)
    plt.show()

    # Plot 4: Distribution of Shape RMSE
    plt.figure(figsize=(10, 6))
    sns.histplot(data['shape_rmse'], kde=True, color='red')
    plt.title('Distribution of Shape RMSE')
    plt.xlabel('Shape RMSE')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

    # Plot 5: Growth Rate Gamma over Time
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='growth_rate_time', y='growth_rate_gamma', data=data, marker='o', color='blue')
    plt.title('Growth Rate Gamma over Time')
    plt.xlabel('Growth Rate Time')
    plt.ylabel('Growth Rate Gamma')
    plt.grid(True)
    plt.show()

    # Get the numerical columns for comprehensive visualization
    numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
    
    # Pair plot to visualize relationships between all numerical features
    sns.pairplot(data[numerical_cols])
    plt.suptitle('Pairwise Relationships', y=1.02)
    plt.show()

    # Correlation heatmap
    plt.figure(figsize=(12, 8))
    correlation_matrix = data[numerical_cols].corr()
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
    plt.title('Correlation Heatmap')
    plt.show()

    # Box plots to visualize distributions and outliers
    for col in numerical_cols:
        plt.figure(figsize=(8, 4))
        sns.boxplot(data[col], color='lightblue')
        plt.title(f'Box Plot of {col}')
        plt.xlabel(col)
        plt.grid(True)
        plt.show()

# Example: Visualize data from MOESM3_ESM.txt
file_path = r'C:\Users\Frikha Rayen\Desktop\LBC\HACKATON Oct 2024\41586_2021_4301_MOESM3_ESM.txt'
data = pd.read_csv(file_path, comment='#', sep=',', on_bad_lines='skip', engine='python')

# Call the visualization function for the full dataset
visualize_full_dataset(data)
```
